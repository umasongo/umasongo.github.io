```python
import pandas as pd
df = pd.read_csv("data-excel.csv")
df
```

```python
df.columns
```

```python
df_melt=pd.melt(df, 
                id_vars=['District','Reg. Voters','Total Votes','Valid Votes','Received Station'],
                value_vars=['AMURIAT OBOI PATRICK','KABULETA KIIZA JOSEPH','KALEMBE NANCY LINDA','KATUMBA JOHN',
                           'KYAGULANYI SSENTAMU ROBERT', 'MAO NORBERT','MAYAMBALA WILLY','MUGISHA MUNTU GREGG',
                           'MWESIGYE FRED','TUMUKUNDE HENRY KAKURUGU','YOWERI MUSEVENI TIBUHABURWA KAGUTA','Invalid Votes'],
                var_name='Candidate',value_name='Candidate votes')

df_melt["District"] = df_melt["District"].str.strip().str.split(' ', n=1, expand=True).iloc[:, 1]

split = df_melt['Total Votes'].str.split('\n', n=1, expand=True)
df_melt["Total Votes"] = split.iloc[:, 0].astype(int)
df_melt["Total Votes percent"] = split.iloc[:, 1].str.replace('%','').astype(float)

split = df_melt['Received Station'].str.split('\n', n=1, expand=True)
df_melt['Received Station'] = split.iloc[:, 0].str.replace('/','').astype(int)
df_melt['Received Station total'] = split.iloc[:, 1].astype(int)

split = df_melt['Candidate votes'].str.split('\n', n=1, expand=True)
df_melt["Candidate votes"] = split.iloc[:, 0].astype(int)
df_melt["Candidate Votes percent"] = split.iloc[:, 1].str.replace('%','').astype(float)


df_melt
```

```python
df_melt.columns = df_melt.columns.str.lower().str.replace(' ', '_')
df_melt.rename(columns = {'reg._voters':'reg_voters'}, inplace = True)
df_melt = df_melt[['district','reg_voters','total_votes','total_votes_percent','valid_votes','received_station','received_station_total','candidate','candidate_votes','candidate_votes_percent']]

df_melt
```
